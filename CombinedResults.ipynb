```python
d={'League':['EPL','NBA','NFL','MLB'],'BestTeamChampionships':[
    sum((EPL_Results.Team==1)&(EPL_Results.Rank==1))/5000,
    sum((NBA_Results.TeamRank==1)&(NBA_Results.PlayoffResult=='Champ'))/5000,
    sum((NFL_Results.TeamRank==1)&(NFL_Results.PlayoffResult=='Champ'))/5000,
    sum((MLB_Results.TeamRank==1)&(MLB_Results.PlayoffResult=='Champ'))/5000,
]}
FirstResults=pd.DataFrame(data=d)
FirstResults
```

```python
plt.rcParams["figure.figsize"] = [14, 5]
fig = plt.figure()
ax = fig.add_axes([.1,.1,.5,.5])
plt.bar(FirstResults.League,FirstResults.BestTeamChampionships)
ax.set_ybound(0,1)
ax.set_xlabel('League')
ax.set_ylabel('Fraction of Championships')
ax.grid(axis='y')
plt.title('Championships Won by Best Team')
plt.show()
```
